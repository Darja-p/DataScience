**Scrape data from HTML tables into dataframe using BeautifulSoup and Pandas**


```python
import pandas as pd
url="https://en.wikipedia.org/wiki/Polynesia"
data = requests.get(url).text
soup = BeautifulSoup(data, "html.parser")
tables = soup.find_all("table")
len(tables)
```

```python
for index,table in enumerate(tables):
    if ("Country or dependency" in str(table)):
        table_index = index
print(table_index)
```

```python
print(tables[table_index].prettify())
```

```python
import re
country_data = pd.DataFrame(columns=["Country of dependency", "Notes"])

for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        country = col[0].text
        note = col[1].text
        country = re.sub(r'\n','',country)
        note = re.sub(r'\n','', note)
        country_data = country_data.append({"Country of dependency":country, "Notes":note }, ignore_index=True)

        
country_data
```
