## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

mpl.style.use('Solarize_Light2') # optional: for ggplot-like style


QUERY = """SELECT ConvertedComp from master"""

df = pd.read_sql_query(QUERY,conn)

df['ConvertedComp'].plot.hist()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """SELECT Age from master"""

df = pd.read_sql_query(QUERY,conn)


df.plot(kind='box', figsize=(8, 6))

plt.title('Box plot of the Age of Survey Participants')
plt.ylabel('Age')
plt.grid()
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """SELECT * from master"""

df = pd.read_sql_query(QUERY,conn)

df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkblue')

plt.title('Hours worked per week and Age')
plt.xlabel('Age')
plt.ylabel('Hours worked per week')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

df.plot(kind='scatter', x='CodeRevHrs', y='WorkWeekHrs', figsize=(10, 6), color='darkblue', alpha = 0.5, s = df['Age'])

plt.title('Hours worked per week and Code Review Hours and Age')
plt.xlabel('Code Review Hours')
plt.ylabel('Hours worked per week')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python

QUERY = """SELECT * from DatabaseDesireNextYear"""

df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
df1 = df.groupby(['DatabaseDesireNextYear']).count()
#mb = list(dfa['MainBranch'].values)
df1.sort_values('Respondent', inplace = True, ascending = False)
df1 = df1.head(5)

df1['Respondent'].plot(kind='barh', figsize=(10, 6))
```

```python
df_db = df.groupby('DatabaseDesireNextYear', axis=0).count()
df_db.sort_values(by=['Respondent'], inplace=True, ascending=False)
df_db= df_db.head()
```

```python
# your code goes here
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
df_db['Respondent'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° 
                            shadow=True,       # add shadow     
                            pctdistance=0.7,
                            colors = colors_list
                            )

plt.title('Top 5 databases that respondents wish to learn next year')
#plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """SELECT * from master"""

df = pd.read_sql_query(QUERY,conn)
age = [30, 31, 32, 33, 34, 35]
df_age = df[df['Age'].isin(age)] 
#df.query('A in [6, 3]')
df_age = df_age[['WorkWeekHrs', 'CodeRevHrs', 'Age']]
df1 = df_age.groupby(['Age'])[['WorkWeekHrs', 'CodeRevHrs']].median()
df1
```

```python
# your code goes here


df1[['WorkWeekHrs', 'CodeRevHrs']].plot(kind='bar', stacked=True)
plt.title('Median of hours worked weekly and hours spent on code review')
plt.show()
```

Plot the median `ConvertedComp` for all ages from 45 to 60.
